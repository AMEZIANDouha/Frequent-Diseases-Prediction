### Logistic Regression 

Nous allons initialiser notre modèle avec des paramètres initiaux que nous allons éssayer d'améliorer par la suite !

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import balanced_accuracy_score



# Initialiser et entraîner le modèle de régression logistique
logistic_regression_model = LogisticRegression(random_state=42,penalty='l2',C=0.1018)

logistic_regression_model.fit(X_train, y_train)


# Faire des prédictions sur l'ensemble de test
y_pred = logistic_regression_model.predict(X_test)

bas = balanced_accuracy_score(y_test, y_pred)
#print(f"Accuracy équilibrée: {bas:.3f}")
```

**random_state=42 : Fixe la graine aléatoire pour garantir la reproductibilité des résultats lors de l'entraînement du modèle de régression logistique.

**penalty='l2' : Applique la régularisation L2 (Ridge) pour prévenir le surajustement en ajoutant un terme de pénalité aux coefficients du modèle.

**C=0.10018 : Contrôle l'influence de la régularisation, avec une valeur plus petite de C indiquant une régularisation plus forte, contribuant à la stabilité du modèle.

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# Initialiser le  modèle
model = DecisionTreeClassifier(random_state=42,min_samples_split=5)


# Entraîner le modèle
model.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_dt = model.predict(X_test)

# Faire des prédictions sur l'ensemble de test
accuracy = balanced_accuracy_score(y_test, y_pred_dt)

print("Précision du modèle :", accuracy)
```

**random_state=42 :** Cela fixe la graine (seed) du générateur de nombres aléatoires. Fixer la graine garantit que le modèle donnera des résultats reproductibles, ce qui est utile lors du développement et du débogage.

**min_samples_split=5 :** C'est le nombre minimum d'échantillons nécessaires pour diviser un nœud interne. Si le nombre d'échantillons dans un nœud est inférieur à min_samples_split, le nœud ne sera pas scindé, ce qui peut aider à prévenir la sur-ajustement (overfitting).

### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

# Initialiser le SVM modèle
model_ =SVC(kernel='rbf',C=0.3,probability=True, random_state=42)

# Entraîner le modèle
model_.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_svm = model_.predict(X_test)

# Faire des prédictions sur l'ensemble de test
accuracy = balanced_accuracy_score(y_test, y_pred_svm)
print("Accuracy:", accuracy)
```

**kernel='rbf': Le noyau radial (RBF) est utilisé pour mesurer la similarité entre les points.                                    
**C=0.3: Paramètre de régularisation, contrôlant la tolérance aux erreurs sur l'ensemble d'entraînement.

### Naive Bayes 

```python
from sklearn.naive_bayes import MultinomialNB


# Initialize the Multinomial Naive Bayes model
model_nb = MultinomialNB(alpha=6)

# Train the model
model_nb.fit(X_train, y_train)

# Make predictions on the test set
y_pred_nb = model_nb.predict(X_test)

# Evaluate the model
accuracy = balanced_accuracy_score(y_test, y_pred_nb)
print("Accuracy:", accuracy)
```

**alpha=6 :** Paramètre de lissage (additif Laplace / Lidstone), qui contrôle la régularisation en ajoutant une petite valeur à toutes les fréquences de caractéristiques, évitant ainsi les probabilités nulles.

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier

# Initialiser et entraîner le modèle Random Forest
model_rf = RandomForestClassifier(random_state=42,max_depth=10)
model_rf.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_rf = model_rf.predict(X_test)

# Évaluer la performance du modèle
accuracy_rf = balanced_accuracy_score(y_test, y_pred_rf)
print("Précision du modèle Random Forest :", accuracy_rf)
```

**Les forêts aléatoires** combinent plusieurs arbres de décision pour obtenir une prédiction robuste et sont souvent utilisées pour des tâches de classification et de régression.

**max_iter=1000 :** Le nombre maximal d'itérations (époques) lors de l'optimisation du modèle. Cela contrôle le nombre de fois que le modèle va apprendre sur l'ensemble de données d'entraînement.

**Accuracy (Précision) :**
Mesure la proportion totale de prédictions correctes. Une accuracy élevée indique une capacité globale du modèle à classer correctement les instances.

**AUC (Area Under the Curve) :**
Représente la capacité du modèle à discriminer entre les classes. Une AUC proche de 1.0 indique une excellente performance de discrimination.

**Recall (Rappel) :**
Indique la capacité du modèle à capturer les instances positives. Un rappel élevé est important lorsque la non-détection de vrais positifs est coûteuse.

**Precision (Précision) :**
Mesure la précision du modèle parmi les instances qu'il a prédites comme positives. Une haute précision est importante lorsque les faux positifs sont coûteux.

**F1 Score :**
Fournit un équilibre entre rappel et précision. Un F1 score élevé est souhaitable lorsque les classes sont déséquilibrées.

**Kappa (Cohen's Kappa) :**
Mesure la concordance ajustée pour les déséquilibres de classe. Un kappa élevé indique une bonne concordance entre les prédictions et les observations réelles.

**MCC (Matthews Correlation Coefficient) :**
Évalue la qualité des classifications binaires. Une valeur élevée indique une forte corrélation entre les prédictions du modèle et les vraies classes.

**Temps d'inférence (TT) :**
Le temps total nécessaire pour effectuer les prédictions. Des temps plus courts peuvent être préférables dans des scénarios où l'efficacité de l'inférence est cruciale.

## Validation Croisée 

### La validation croisée offre une estimation plus fiable de la performance du modèle qu'une seule division entraînement-test et donne également une idée sur la possibilité d'amélioration des hyperparamètres si les résultats obtenues par validation croisée montrent un écart significatif par rapports aux résultats déjà obtenues.

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

num_folds = 7 

# Specify the cross-validation strategy (StratifiedKFold for classification)
cv_strategy = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

# Loop through the models and perform cross-validation
for model_name, model in models:
    # Perform cross-validation
    cross_val_results = cross_val_score(model, X, y, cv=cv_strategy, scoring="balanced_accuracy")  
    # Display cross-validation results
    print(f"\nModel: {model_name}")
    print(f"Cross-Validation Accuracy: {cross_val_results}")
    print(f"Mean Accuracy: {cross_val_results.mean()}")
    
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold


# Créer un DataFrame vide pour stocker les résultats
results_df = pd.DataFrame(columns=['Model', 'Mean Accuracy', 'Standard Deviation'])

# Loop through the models and perform cross-validation
for model_name, model in models:
    # Perform cross-validation
    cross_val_results = cross_val_score(model, X, y, cv=cv_strategy, scoring="balanced_accuracy")  
    
    # Ajouter les résultats au DataFrame
    results_df = pd.concat([results_df, pd.DataFrame({
        'Model': [model_name],
        'Mean Accuracy': [cross_val_results.mean()],
        'Standard Deviation': [cross_val_results.std()]
    })], ignore_index=True)

# Afficher le DataFrame
results_df
```

## Cross-tuning 

### Le "cross tuning" (ou "grid search") consiste à trouver les meilleurs hyperparamètres d'un modèle en évaluant sa performance sur différentes combinaisons prédéfinies d'hyperparamètres, mesurée à l'aide de la validation croisée. Cela permet d'optimiser la configuration du modèle pour obtenir les meilleures performances sur des données non vues, améliorant ainsi sa capacité à généraliser à de nouveaux exemples.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

models = [
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Neural Network', MLPClassifier(random_state=42, max_iter=1000))  
]

param_grids = [
    {'criterion': ['gini', 'entropy'], 'max_depth': [None, 10, 20, 30], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4]},
    {'n_estimators': [80, 100, 200], 'max_depth': [None, 10, 20, 40], 'min_samples_split': [3, 5, 10], 'min_samples_leaf': [3, 2, 4]},
    {'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)], 'activation': ['logistic', 'tanh', 'relu'], 'alpha': [0.0001, 0.001, 0.01], 'learning_rate': ['constant', 'invscaling', 'adaptive']}
]

# Perform grid search for each model
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for (model_name, model), param_grid in zip(models, param_grids):
    grid_search = GridSearchCV(model, param_grid, cv=cv_strategy, scoring='accuracy')
    grid_search.fit(X, y)

    best_params = grid_search.best_params_
    print(f"Best parameters for {model_name}: {best_params}")

    # Evaluate the model with the best parameters using cross-validation
    model.set_params(**best_params)  # Set the best parameters
    pipeline = make_pipeline(StandardScaler(), model)  # Include scaling for some models
    cv_scores = cross_val_score(pipeline, X, y, cv=cv_strategy, scoring='balanced_accuracy')
    print(f"Cross-validation accuracy for {model_name}: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})\n")
```
